```python
import numpy as np
```

```python
import seaborn as sns
```

```python
sns.set_style('ticks')
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
z = np.linspace(-5, 5, 1000)
```

```python
z1 = np.linspace(-6, -5, 10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(z, sigmoid(z), linewidth=2, zorder=3, color='crimson')
plt.plot(z1, sigmoid(z1), linewidth=2, color='crimson', linestyle='dashed')
plt.plot(-z1, sigmoid(-z1), linewidth=2, color='crimson', linestyle='dashed')
plt.axvline(x=0, linewidth=1, linestyle='dashed', color='black')
plt.xlabel('z', fontsize=14)
plt.ylabel('Sigmoid(z)', fontsize=14)
plt.savefig('sigmoid.png', dpi=300)
```

```python
h = np.linspace(0, 1, 1000)
```
